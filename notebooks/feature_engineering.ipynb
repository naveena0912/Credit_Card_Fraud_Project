```python
import numpy as np

df["hour_mod24"] = ((df["Time"]//3600) % 24).astype(int)
df["is_night"] = df["hour_mod24"].isin([0,1,2,3,4,5,22,23]).astype(int)
df["is_business_hours"] = df["hour_mod24"].between(9, 17).astype(int)

df["log_amount"] = np.log1p(df["Amount"])

print(df["is_night"].value_counts())
print(df["is_business_hours"].value_counts())
print(df["log_amount"].head())
```
